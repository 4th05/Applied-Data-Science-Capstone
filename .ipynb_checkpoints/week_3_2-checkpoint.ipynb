```python
import pandas as pd
import geocoder
```

```python
df = pd.read_pickle("List_of_postal_codes_of_Canada.pkl")
```

```python
lats = [geocoder.arcgis('{}, Toronto, Ontario'.format(ps)).latlng for ps in df["Postal Code"]]
```

```python
df["Latitude"] = list(list(zip(*lats))[0])
df["Longitude"] = list(list(zip(*lats))[1])
```
